```python
from datetime import datetime as dt
import sqlalchemy
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, Text, ForeignKey, DateTime, create_engine
from sqlalchemy.orm import sessionmaker, relationship
Base = declarative_base()


class Game(Base): #child
    __tablename__ = 'games'
    id = Column(Integer, primary_key = True)
    venue = Column(Text)
    date = Column(DateTime)    
    home_team_id = Column(Integer,ForeignKey('home_teams.id'))
    home_team = relationship('HomeTeam', back_populates='games')
    away_team_id = Column(Integer,ForeignKey('away_teams.id'))
    away_team = relationship('AwayTeam', back_populates='games')
#     winner = Column(Text) #Kyle's suggestion is to make winner an argument of the Game class instead of tied to a winner
# #     class
    winner_id = Column(Integer,ForeignKey('winners.id'))
    winner = relationship('Winner', back_populates='games')

# originally tried to make the Team the only class with relationship to game. tried it as both the parent and child,
# but the duplication of names and objects was throwing ev off:
# class Team(Base): #parent
#     __tablename__ = 'teams'
#     id = Column(Integer, primary_key = True)
#     country = Column(Text)
#     games = relationship('Game', back_populates='team')
        

    
class HomeTeam(Base): #parent
    __tablename__ = 'home_teams'
    id = Column(Integer, primary_key = True)
    country = Column(Text)
    games = relationship('Game', back_populates='home_team')
    
class AwayTeam(Base): #parent
    __tablename__ = 'away_teams'
    id = Column(Integer, primary_key = True)
    country = Column(Text)
    games = relationship('Game', back_populates='away_team')
    
class Winner(Base): #parent
    __tablename__ = 'winners'
    id = Column(Integer, primary_key = True)
    country = Column(Text)
    games = relationship('Game', back_populates='winner')
```
